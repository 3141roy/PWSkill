#### Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a micro web framework for Python that allows developers to build web applications quickly and efficiently. Flask is lightweight, flexible, and easy to learn, making it a popular choice for building small to medium-sized web applications.

Advantages of Flask Framework:

1. Lightweight and Flexible: Flask is a lightweight framework that doesn't impose any specific architecture or design patterns. It allows developers to build web applications in any way they prefer, giving them the flexibility to choose their own approach.

2. Easy to Learn: Flask is easy to learn and doesn't require much prior knowledge of web development or Python. Its simple and intuitive API makes it a good choice for beginners.

3. Extensible: Flask is highly extensible and allows developers to add new features and functionality easily through its modular design.

4. Integrates Well with Other Technologies: Flask integrates well with other technologies such as SQLAlchemy, Jinja2, and WTForms, making it a powerful tool for building web applications.

5. Good Documentation and Community Support: Flask has good documentation and an active community that provides support and resources for developers.

6. Scalability: Flask applications can be easily scaled up or down as needed, making it a suitable framework for building applications of any size.

7. Testing: Flask has built-in support for testing, making it easy to test the application and catch bugs early in the development process.

#### Q3. What is App routing in Flask? Why do we use app routes?

App routing in Flask refers to the process of mapping URLs to view functions in a Flask application. Flask uses a decorator called @app.route to define routes and associate them with view functions.

When a user requests a URL that matches a defined route in the Flask app, the corresponding view function is executed and its return value is sent back to the user as a response. Flask supports various HTTP methods like GET, POST, PUT, DELETE, etc., and we can specify which HTTP methods a route should respond to using the methods argument in the @app.route decorator.
