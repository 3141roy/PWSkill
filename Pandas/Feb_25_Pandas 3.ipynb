#### Q2. What is the difference between the functions loc and iloc in pandas.DataFrame?

loc is a label-based indexing method that is used to access rows and columns by labels or boolean arrays. The syntax of the loc method is df.loc[row_labels, column_labels]. Here, row_labels and column_labels refer to the row and column labels, respectively. loc returns a DataFrame that matches the specified labels.

iloc on the other hand, is an integer-based indexing method that is used to access rows and columns by integer position. The syntax of the iloc method is df.iloc[row_positions, column_positions]. Here, row_positions and column_positions refer to the integer positions of the rows and columns, respectively. iloc returns a DataFrame that matches the specified integer positions.

#### Q4. Write a code to find the following statistical measurements for the above dataframe df1:
(i) mean of each and every column present in the dataframe.
(ii) standard deviation of column, column_2

```python
import numpy as np
columns = ['column_1', 'column_2', 'column_3', 'column_4', 'column_5', 'column_6']
indices = [1,2,3,4,5,6]
#Creating a dataframe:
df1 = pd.DataFrame(np.random.rand(6,6), columns = columns, index = indices)

means=[]

for each_col in list(df1.columns):
    means.append(df1[each_col].mean())

print(f"Mean of each column: {means}")
print(f"Standard Deviation of column_2: {df1['column_2'].std()}")
```

#### Q6. What do you understand about the windows function in pandas and list the types of windows functions?

In pandas, a window function is a function that operates on a set of rows (a window) within a DataFrame. The window is defined by a specific set of rows, and the function is applied to that set of rows to produce a new value for each row in the window.

There are several types of window functions in pandas:

1. Rolling: The rolling function creates a rolling window of a specific size over the DataFrame, and then applies a function to each window.

2. Expanding: The expanding function computes a statistic for the entire dataset, then computes the same statistic for an expanding window of the dataset.

3. Exponentially Weighted: The exponentially weighted function applies exponentially decreasing weights to the values in the window, giving more weight to recent values.

4. Rolling with time intervals: The rolling function can also be used with time intervals, creating a rolling window over a specific time period.

5. Groupby: The groupby function groups the rows in the DataFrame by a specific column, and then applies a function to each group.

6. Window: The window function applies a function to a window defined by a specific set of rows, without any overlap between the windows.

#### Q10. Write a Python program that reads a CSV file containing sales data for different products and visualizes the data using a stacked bar chart to show the sales of each product category over time. The program should prompt the user to enter the file path and display the chart.

```python
filepath = input('Enter filepath: ')

df = pd.read_csv(filepath)
df['Date'] = pd.to_datetime(df['Date'])
grouped = df.groupby(['Product', 'Date']).sum()
pivot = grouped.unstack(level=0)
pivot.plot(kind='bar', stacked=True)
```
